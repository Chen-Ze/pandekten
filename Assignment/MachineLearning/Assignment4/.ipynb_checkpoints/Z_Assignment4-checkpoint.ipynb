```python
# Functions for packing and unpacking images. 
# These are provided.

def packcw(A,nr,nc):
    x = (A.T).reshape(nr*nc,1)
    return x

def unpackcw(x,nr,nc):
    A = x.reshape(nc,nr)
    return A.T

def packrw(A,nr,nc):
    x = A.reshape(nr*nc,1)
    return x

def unpackrw(x,nr,nc):
    A = x.reshape(nr,nc)
    return A
```

```python
# Load MNIST train and test data
# Normalize image pixels to [0,1]

train = np.load('MNISTcwtrain1000.npy').astype(float)/255
test = np.load('MNISTcwtest100.npy').astype(float)/255

size = train.shape[0]
n_train = train.shape[1]
n_test = test.shape[1]

print('-----MNIST Dataset Loaded-----')
print('Train Data: %d x %d' %(size,n_train))
print('Test Data: %d x %d' %(size,n_test))
```

```python
# This function will return the indices of the pixels in 
# a centered square of side length: d = 2, 4, 6, 8,..., 12, 14. 
# It is provided.

def return_mask_idx(sl):
    s = 14-sl//2   
    idx = []
    for i in range(sl):
        for j in range(sl):
            idx.append(28*(s+i)+s+j) 
    return idx
```

### 2. Classify the Images using the nearest mean classifier.

The nearest mean classifier uses the image labels (0,...,9) to compute the mean image of each class ($\mu_0$, $\mu_1$, ...,$\mu_9$). Then classifies a new example to be the label of the nearest class mean.  
$$ \hat y (x) = \arg\min_{i\in [0,9]} \|x- \mu_i\|^2 $$  

(a) Compute the class means using the unmasked training images.   
(b) Classify the training images using the nearest mean classifer. Record the accuracy.  
(c) Classify the testing images using the nearest mean classifer. Record the accuracy. 
(d) For $d=2,..., 14$ classify the masked training images and the masked testing images using the nearest mean classifer.  
(e) Plot the classification error as a function of $d$. On the plot also indicate the classification error for the unmasked images and chance accuracy. Use a legend to distinguish the lines.

```python
# Use the training data to compute the class means
class_means = np.zeros((train.shape[0],10))
for i in range(10):
  class_means[:,i]=(np.mean(train[:,i*1000:(i+1)*1000],axis=1))

def accuracy(data,num):
  classifier_estimate = np.zeros((1,data.shape[1]))

  for i in range(data.shape[1]):
    distance = []
    for j in range(10):
      distance.append(np.linalg.norm(data[:,i]-class_means[:,j]))
    classifier_estimate[0,i] = np.argmin(distance)
  
  return (classifier_estimate.reshape((10,num)).transpose()-np.arange(10)==0).mean()

# Define nearest mean classifier
# First: Classify the training data
accuracy_train = accuracy(train,1000)
print("accuracy of train data:",accuracy_train)
# plt.imshow(unpackcw(train[:,1000], 28, 28), plt.cm.gray)
# Second: classify the testing data
accuracy_test = accuracy(test,100)
print("accuracy of test data:",accuracy_test)

# Third: Classify the masked training data vs d
error_train_masked = []
error_test_masked = []
for dimension in np.arange(2,16,2):
  masked_row = return_mask_idx(dimension)
  train_masked = hide(train, masked_row)
  test_masked = hide(test, masked_row)
  error_train_masked.append(1 - accuracy(train_masked,1000))
  error_test_masked.append(1 - accuracy(test_masked, 100))

# Fourth: Classify the corrupted testing data vs d
plt.figure()
dimension_display = np.arange(2,16,2)
plt.plot(dimension_display,error_train_masked,label='masked train data')
plt.plot(dimension_display,error_test_masked,label='masked test data')
plt.plot(dimension_display,np.ones(7)*(1-accuracy_train),label='unmasked train data')
plt.plot(dimension_display,np.ones(7)*(1-accuracy_test),label='unmasked test data')
plt.plot(dimension_display,np.ones(7)*(0.9),label='chance error')
plt.legend()
plt.xlabel('d')
plt.ylabel('Error')
plt.show
```

### 3. Use least squares to approximate the masked pixels (estimation of the hidden part of the image)
Calculate the MSE for each example by constructing the following affine MSE estimator.

\begin{equation}
\arg \min_{W\in\mathbb{R}^{d^2 x(784-d^2)}}\|y-Wx-b\|_2^2
\end{equation}
\begin{equation}
y \in \mathbb{R}^{d^2}  \hspace{5mm} b \in \mathbb{R}^{d^2} \hspace{5mm} X \in \mathbb{R}^{(784-d^2)}
\end{equation}

Here $y$ is the vector of masked pixel valaues and $x$ is the remainder of the pixel values. 

Then use thr NM classifier to classify each reconstructed test image and each reconstructed traing image.
Plot the accuracies in each on a copy of the previous plot.

```python
# Predict of missing pixels on testing data then classify
def estimate(dimension, data):
  hide_index = return_mask_idx(dimension)
  train_x = np.delete(train, hide_index, axis=0)
  train_y = np.delete(train, np.delete(np.arange(0, 784), hide_index), axis=0)
  W_star, b = solve_ls_affine(train_x, train_y)
  b_m =np.zeros((b.shape[0],1))
  b_m[:,0] = b
  b_matrix = np.tile(b_m,(1,data.shape[1]))
  # print(b_matrix.shape)
  input_x = np.delete(data, hide_index, axis=0)
  estimate_y = W_star.transpose()@ input_x + b_matrix
  # substitute the masked pixels in the input picture
  figure_estimated = np.copy(data)
  figure_estimated[hide_index]=estimate_y

  return figure_estimated
```

```python
# Predict of missing pixels on training data then classify
error_train_estimated = []
error_test_estimated = []
for dimension in np.arange(2,16,2):
  train_estimated = estimate(dimension, train)
  test_estimated = estimate(dimension, test)
  error_train_estimated.append(1 - accuracy(train_estimated,1000))
  error_test_estimated.append(1 - accuracy(test_estimated, 100))
```

```python
  # Plot performance of all 
plt.figure()
dimension_display = np.arange(2,16,2)

plt.plot(dimension_display,error_train_masked,label='masked train data')
plt.plot(dimension_display,error_test_masked,label='masked test data')
plt.plot(dimension_display,np.ones(7)*(1-accuracy_train),label='unmasked train data')
plt.plot(dimension_display,np.ones(7)*(1-accuracy_test),label='unmasked test data')
plt.plot(dimension_display,np.ones(7)*(0.9),label='chance error')

plt.plot(dimension_display,error_train_estimated,label='estimated train data' )
plt.plot(dimension_display,error_test_estimated,label='estimated test data')
plt.legend()
plt.xlabel('d')
plt.ylabel('Error')
plt.show
```
